```python
import mbuild as mb
import numpy as np
from grits import backmap
from grits import CG_Compound
```

```python
bto = mb.load("../grits/tests/assets/bto.mol2")

bto_colors = {"_A1": "blue", "_A2": "yellow", "_D": "red", "_S_L1": "green", "_S_L2":"pink", "_S_S":"black"}
```

```python
cg_beads = {"_A1": "s1nc2c(n1)cccc2",
               "_A2": "C(=C(C#N)C#N)1c2c(C(=O)C(=C)1)cc(F)c(F)c2", 
               "_D": "c23c(c(n)cs2)scc3",
               "_S_S": "CCOCCOCCOC",
               "_S_L1": "CCCCCC",
               "_S_L2": "CCCCC"
               }

cg_bto = CG_Compound(bto, cg_beads)

cg_bto.visualize(color_scheme=bto_colors, show_atomistic=True).show()
cg_bto.visualize(color_scheme=bto_colors).show()
```

```python
cg_bto.save("cg-bto.gsd", overwrite=True)
```
